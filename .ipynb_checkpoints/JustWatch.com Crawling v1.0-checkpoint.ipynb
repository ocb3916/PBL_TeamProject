```python
# 모듈 임포트
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import time
import requests as re
```

```python
# Chrome WebDriver 초기화 (자동으로 ChromeDriver 다운로드 및 설정)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 사이트 접속
url = 'https://www.justwatch.com/kr'  # Replace with the actual URL
driver.get(url)

time.sleep(5)

# 개인정보 관련 동의 버튼 클릭 Personal Information -> pi
# shadow-DOM으로인해 웹사이트 요소에 접근하지 못하는 문제 해결
# https://blog.hashscraper.com/hidden-web-elements-shadow-dom-crawling/
def get_shadow_root(element, driver):
    return driver.execute_script('return arguments[0].shadowRoot', element)

root1 = driver.find_element(By.XPATH, '//*[@id="usercentrics-root"]')
shadow_root1 = get_shadow_root(root1, driver)
pi_button = shadow_root1.find_element(By.CSS_SELECTOR, '#uc-center-container > div.sc-eBMEME.hasieq > div > div > div > button.sc-dcJsrY.gDlDbR')
pi_button.click()

# 인기 미디어 요소들의 HTML을 리스트에 저장
media_elements_html = []
div_num = 2

# 요소들을 탐색하며 HTML 추출
while True:
    try:
        # XPath를 이용해 div[숫자] 형태로 각 요소의 HTML을 추출
        element_xpath = f'//*[@id="base"]/div[3]/div[2]/div/div[1]/div/div[{div_num}]/a'
        element = driver.find_element(By.XPATH, element_xpath)

        # 요소의 HTML을 가져와 리스트에 저장
        element_html = element.get_attribute('href')
        media_elements_html.append(element_html)

        print(f"Element {div_num - 1} URL collected.")
        div_num += 1

    except Exception as e:
        # 요소를 찾을 수 없으면 스크롤을 내림
        print(f"Element {div_num - 1} not found. Scrolling down...")
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(1)  # 스크롤 후 대기
        
        # 스크롤 후 다음 요소를 찾기 전에 div_num에 해당하는 요소가 실제로 있는지 확인
        new_element_xpath = f'//*[@id="base"]/div[3]/div[2]/div/div[1]/div/div[{div_num}]/a'
        new_element = driver.find_elements(By.XPATH, new_element_xpath)

        # 스크롤 후에도 더 이상 새로운 요소가 나타나지 않으면 종료
        if not new_element:
            print("No more new elements found. Exiting.")
            break

print(media_elements_html)
```

```python
# 수집한 각 요소 HTML을 BeautifulSoup으로 분석하여 원하는 정보 추출
for idx, url in enumerate(media_elements_html):
    try:
        time.sleep(1)
        
        # HTTP 요청 보내기 및 BeautifulSoup 객체 생성
        response = re.get(url, headers={'User-agent':'Mozilla/5.0'})
        soup = BeautifulSoup(response.text, 'html.parser')

        # 각 요소의 HTML에서 타이틀 정보 및 서비스 플랫폼 등 필요한 정보 추출
        title_info = soup.find('div', class_='title-info title-info')
        title_info1 = soup.find('div', class_='title-sidebar__title-with-poster__title-text')
        service_platform = soup.find(class_='spinning-texts')
        synopsis = soup.select('#synopsis')

        print(f"\n=== Element {idx + 1} ===")

        # 타이틀 정보 출력 (title_info)
        if title_info:
            for item in title_info:
                print(item.get_text())
        else:
            print("Title info가 존재하지 않습니다.")

        # 타이틀 정보 출력 (title_info1)
        if title_info1:
            for item in title_info1:
                print(item.get_text())
        else:
            print("Title info1이 존재하지 않습니다.")

        # 서비스 플랫폼 정보 출력
        if service_platform:
            for item in service_platform:
                print(item.get_text())
        else:
            print("서비스 플랫폼 정보가 존재하지 않습니다.")

        # 시놉시스 정보 출력
        if synopsis:
            for item in synopsis:
                print(item.get_text())
        else:
            print("시놉시스 정보가 존재하지 않습니다.")

    except Exception as e:
        print(f"Element {idx + 1}에서 오류 발생: {e}")

# 크롤링 종료 후 브라우저 닫기
driver.quit()
```
