```python
# Î™®Îìà ÏûÑÌè¨Ìä∏
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import time
import requests as re
```

```python
# Chrome WebDriver Ï¥àÍ∏∞Ìôî (ÏûêÎèôÏúºÎ°ú ChromeDriver Îã§Ïö¥Î°úÎìú Î∞è ÏÑ§Ï†ï)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# ÏÇ¨Ïù¥Ìä∏ Ï†ëÏÜç
url = 'https://www.justwatch.com/kr'  # Replace with the actual URL
driver.get(url)
```

```python
# Í∞úÏù∏Ï†ïÎ≥¥ Í¥ÄÎ†® ÎèôÏùò Î≤ÑÌäº ÌÅ¥Î¶≠ Personal Information -> pi
# shadow-DOMÏúºÎ°úÏù∏Ìï¥ ÏõπÏÇ¨Ïù¥Ìä∏ ÏöîÏÜåÏóê Ï†ëÍ∑ºÌïòÏßÄ Î™ªÌïòÎäî Î¨∏Ï†ú Ìï¥Í≤∞
# https://blog.hashscraper.com/hidden-web-elements-shadow-dom-crawling/
def get_shadow_root(element, driver):
    return driver.execute_script('return arguments[0].shadowRoot', element)

root1 = driver.find_element(By.XPATH, '//*[@id="usercentrics-root"]')
shadow_root1 = get_shadow_root(root1, driver)
pi_button = shadow_root1.find_element(By.CSS_SELECTOR, '#uc-center-container > div.sc-eBMEME.hasieq > div > div > div > button.sc-dcJsrY.gDlDbR')
pi_button.click()
```

```python
# ÎØ∏ÎîîÏñ¥ Î™©Î°ù Ï§ë ÌïòÎÇò ÌÅ¥Î¶≠
media_button = driver.find_element(By.XPATH, '//*[@id="base"]/div[3]/div[2]/div/div[1]/div/div[2]/a/span/span')
media_button.click()
```

```python
# Í≤∞Í≥º Ï£ºÏÜå Í∞ÄÏ†∏Ïò§Í≥† ÏÖÄÎ†àÎãàÏõÄ Ï¢ÖÎ£å
results_html = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(results_html, 'html.parser')

title_info = soup.find('div', class_='title-info title-info')
title_info1 = soup.find('div', class_='title-sidebar__title-with-poster__title-text')
service_platform = soup.find(class_='spinning-texts')
synopsis = soup.select('#synopsis')


for item in title_info:
    print(item.get_text())
    
for item in title_info1:
    print(item.get_text())
    
for item in service_platform:
    print(item.get_text())

for item in synopsis:
    print(item.get_text())
```
