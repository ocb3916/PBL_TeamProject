```python
# 모듈 임포트
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import time
import requests as re
```

```python
# Chrome WebDriver 초기화 (자동으로 ChromeDriver 다운로드 및 설정)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 사이트 접속
url = 'https://www.justwatch.com/kr'  # Replace with the actual URL
driver.get(url)
```

```python
# 개인정보 관련 동의 버튼 클릭 Personal Information -> pi
# shadow-DOM으로인해 웹사이트 요소에 접근하지 못하는 문제 해결
# https://blog.hashscraper.com/hidden-web-elements-shadow-dom-crawling/
def get_shadow_root(element, driver):
    return driver.execute_script('return arguments[0].shadowRoot', element)

root1 = driver.find_element(By.XPATH, '//*[@id="usercentrics-root"]')
shadow_root1 = get_shadow_root(root1, driver)
pi_button = shadow_root1.find_element(By.CSS_SELECTOR, '#uc-center-container > div.sc-eBMEME.hasieq > div > div > div > button.sc-dcJsrY.gDlDbR')
pi_button.click()
```

```python
# 미디어 목록 중 하나 클릭
media_button = driver.find_element(By.XPATH, '//*[@id="base"]/div[3]/div[2]/div/div[1]/div/div[2]/a/span/span')
media_button.click()
```

```python
# 결과 주소 가져오고 셀레니움 종료
results_html = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(results_html, 'html.parser')

title_info = soup.find('div', class_='title-info title-info')
title_info1 = soup.find('div', class_='title-sidebar__title-with-poster__title-text')
service_platform = soup.find(class_='spinning-texts')
synopsis = soup.select('#synopsis')

for item in title_info:
    print(item.get_text())
    
for item in title_info1:
    print(item.get_text())
    
for item in service_platform:
    print(item.get_text())

for item in synopsis:
    print(item.get_text())
```
